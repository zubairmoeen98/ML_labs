## Lab Overview
This notebook performs outlier detection on the cleaned AEP time-series dataset. The code uses a boxplot and the Interquartile Range method to identify unusual AEP_MW values.

## Learning Objectives

- Visualize the AEP_MW column using a boxplot.

## Section 1: Library Import and Dataset Loading
This section imports the required libraries and loads the dataset after missing value treatment.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'Z:\University\8th Semester\ML Lab\Lab 3\2_Missing_Values_Filled.csv', parse_dates=['Datetime'], index_col='Datetime')
```

```python
df.boxplot('AEP_MW')
```

```python
def detect_outliers_iqr(data):
    outliers = []
    #data = sorted(data)
    q1 = data['AEP_MW'].quantile(0.25)
    q3 = data['AEP_MW'].quantile(0.75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    idx = 2
    idx_list = []
    for i in list(data['AEP_MW']): 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
            idx_list.append(idx)
        idx = idx + 1
    data.loc[data['AEP_MW'] < lwr_bound, 'AEP_MW'] = np.nan
    data.loc[data['AEP_MW'] > upr_bound, 'AEP_MW'] = np.nan
    return outliers, data, idx_list# Driver code
```

## Section 2: Visual Outlier Checking
The boxplot is used to visually inspect the distribution of AEP_MW values and possible outliers.


```python
sample_outliers, p_data, index_of_out = detect_outliers_iqr(df)
print("Outliers from IQR method: ", sample_outliers)
```
