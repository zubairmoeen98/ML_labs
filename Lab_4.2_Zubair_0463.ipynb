## Section 1: Library Import and Feature Dataset Loading
This section imports the required libraries and loads the feature-engineered dataset using Datetime as the index.


```python
import pandas as pd
import numpy as np
#from dataprep.datasets import load_dataset   # Used to import / load dataset
#from dataprep.eda import create_report   # Used to load Report formate
```

```python
df=pd.read_csv(r'Z:\University\8th Semester\ML Lab\Lab 4\5_features_extracted.csv', index_col=['Datetime'], parse_dates=['Datetime'])
df.head()
```

## Section 2: Correlation Analysis with Target Variable
The following cells calculate Pearson, Kendall, and Spearman correlations to understand how each feature is related to AEP load.


```python
print('pearson\n\n',df.corrwith(df["aep"],method="pearson"))
```

```python
print('kendall\n\n',df.corrwith(df["aep"],method="kendall"))
```

```python
print('spearman\n\n',df.corrwith(df["aep"],method="spearman"))
```
